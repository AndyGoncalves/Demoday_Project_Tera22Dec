```python
df = pd.read_csv("/content/drive/MyDrive/Tera_projetoDemoday/Entregas/CAGEDMOV202207.txt",  
            sep = ';', # the separator of columns (generally is ',' or ';')
            header = 0, # header = line with the labels of columns
            decimal = '.', # python works with '.' as decimal separator 
            encoding='UTF-8', # or 'Latin-1' for some special characters: ~ ´ ` ç
            )
ocup = pd.read_excel('/content/drive/MyDrive/Tera_projetoDemoday/Entregas/layout_caged.xlsx', sheet_name='cbo2002ocupação')
```

MERGE

```python
data = df.merge(ocup, # which dataframe to merge
                             how = 'inner', # how to merge the dfs: 'inner', 'outer', 'left', 'right'                    
                             left_on = 'cbo2002ocupação', # what is the label of key in the left dataframe
                             right_on = 'Código' # what is the label of key in the right dataframe
                             ).drop(columns = ['Código'])
```

```python
data.columns
```

```python
data.head(3)
```

FILTER

```python
# Filtering (uf = SP | city = São Paulo | CBO = 212 [Aréas relacionadas a tecnologia])
df1 = data[(data.uf == 35) & (data.município == 355030) & (data.cbo2002ocupação.isin([212205, 212210, 212215, 212305, 212310, 212315, 212320, 212405, 212410, 212415, 212420, 212425, 212430]))]
df1.head(3)
```

DROP

```python
df1 = df1.drop(columns=['competênciamov', 'região', 'uf','município', 'competênciadec'])
```

CHANGING LABELS

```python
df1.columns
```

```python
df1.columns = ['secao', 'subclasse', 'saldomov', 'cbo2002',
       'categoria', 'grau', 'idade', 'horascontratuais', 'racacor',
       'sexo', 'tipoempregador', 'tipoestabelecimento', 'tipomovimentacao',
       'tipodedeficiencia', 'indtrabintermitente', 'indtrabparcial', 'salario',
       'tamestabjan', 'indicadoraprendiz', 'origemdainformação',
       'indicadordeforadoprazo', 'unidadesalariocodigo', 'valorsalariofixo',
       'nome_ocu']
```
